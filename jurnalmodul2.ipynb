```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Assets/aktivis.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
aktivis = plt.imread("Assets/aktivis.jpg")

panel_width = aktivis.shape[1] // 6
panels = [aktivis[:, i * panel_width:(i + 1) * panel_width, :] for i in range(6)]

new_order = [1, 5, 3, 0, 4, 2]
reordered_img = np.concatenate([panels[i] for i in new_order], axis=1)

plt.imshow(reordered_img)
plt.axis('off')
plt.title('The Timeline')
plt.show()

plt.imsave("hasil_urut.png",reordered_img)
```
